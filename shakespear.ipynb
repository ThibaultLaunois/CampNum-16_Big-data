```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
import pyspark.sql.functions as sf
import pyspark
import numpy as np
```

```python
sc = SparkContext("local", "test")
spark = SparkSession.builder.appName("test").getOrCreate()
```

# Load Data

```python
textFile = sc.textFile("beautiful_stories.txt",6)
```

# Most Words

```python
wordCounts = textFile.flatMap(lambda x: x.lower().split()).map(lambda x: x.strip('“".?:,;.-:')).countByValue()
```

```python
wordCounts = sorted(
        wordCounts.items(),
        key=lambda x:x[1],
        reverse=True
    )
```
