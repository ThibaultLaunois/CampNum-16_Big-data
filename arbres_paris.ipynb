# Arbre le plus vieux

```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
import pyspark
```

```python
sc = SparkContext("local", "test")
spark = SparkSession.builder.appName("test").getOrCreate()
```

```python
df = spark.read.csv("arbresremarquablesparis.csv", sep=";", header=True)
```

```python
df.show()
```

```python
old_tree = 2025-int(df.agg({"Année de plantation":"min"}).collect()[0][0])

print(f"Le plus vieil arbre a {old_tree} ans.")
```
